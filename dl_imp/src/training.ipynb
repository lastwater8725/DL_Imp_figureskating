```python
import pandas as pd
from glob import glob
```

```python
data1 = pd.read_csv('./data/jump/output_2/Motion1-4_pose_landmarks.csv')
data1.head()
```

```python
jump_files = glob("./data/jump/output_2/*.csv")
jump_data = pd.DataFrame()
```

```python
for jump_file in jump_files:
    temp = pd.read_csv(jump_file)
    jump_data = pd.concat([temp, jump_data])

jump_data.to_csv("./data/jump/output_2/jump_data.csv")
```

```python
jump_data.head()
```

```python
jump_data.info
```

```python
spin_files = glob("./data/spin/output_2/*.csv")
spin_data = pd.DataFrame()
```

```python
for spin_file in spin_files:
    temp = pd.read_csv(spin_file)
    spin_data = pd.concat([spin_data, temp])

spin_data.to_csv("./data/spin/output_2/*.csv")
```

```python
spin_data.head()
```

```python
spin_data.info()
```

```python
step_files = glob("./data/step/output_2/*.csv")
step_data = pd.DataFrame()
```

```python
for step_file in step_files:
    temp = pd.read_csv(step_file)
    step_data = pd.concat([step_data, temp])

step_data.to_csv("./data/step/output_2/*.csv")
```

```python
step_data.head()
```

```python
step_data.info()
```

```python
total_data = pd.concat([step_data, spin_data, jump_data], axis=0)

total_data.to_csv("./data/total/*.csv")
```
